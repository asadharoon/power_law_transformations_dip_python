# Import Libraries

```python
from skimage import data
from scipy.misc import imread, imresize
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

# Auto Contrast Transformation

Formula of Linear Point Transformation is:
s =(L-1)(r-rmin)/(rmax-rmin)

# Original Image

```python
original_image =imread('salzburg_before.png',True, 'L')    #read image as grey scale image
```

# Processed Image

```python
processed_image =original_image.copy()
```

```python
#s =(L-1)(r-rmin)/(rmax-rmin)
img=processed_image.copy()
rmin=np.amin(processed_image)
rmax=np.amax(processed_image)
processed_img=img.astype(np.uint8)
processed_img=(255)*(img-rmin)/(rmax-rmin)
```

```python
#Normalizing Intensity Levels
processed_image[processed_image<0]=0
processed_image[processed_image>256]=256
```

```python
fig, axes = plt.subplots(1, 2)
ax = axes.ravel()

ax[0].imshow(original_image,cmap=plt.cm.gray,interpolation='bilinear')
ax[0].set_title("Original")
ax[1].imshow(processed_img, cmap=plt.cm.gray,interpolation='bilinear')
ax[1].set_title("Auto Contrast")
plt.imshow(processed_image,cmap=plt.cm.gray) 
```
